```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    itr=1000
    n = len(y)
    lr=0.001
    for i in range(itr):
        yp=m_curr*x+b_curr
        mse=sum((y-yp)**2)/n
        md= (2/n)* sum(-x *(y-yp))
        bd=(2/n)* sum(-(y-yp))
        m_curr=m_curr-lr*md
        b_curr=b_curr-lr*bd
        print(m_curr, b_curr,i)
    return m_curr,b_curr

```

```python
gradient_descent(x,y)
```
